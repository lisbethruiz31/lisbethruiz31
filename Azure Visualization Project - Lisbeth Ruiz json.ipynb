> This dataset explores New York City Airbnb listings from 2008-2015.

## **Number of Airbnb Listings in Each Borough**

```sql
SELECT Borough, COUNT(*) AS Listing_Count
FROM [AirBnB in NYC]
GROUP BY Borough
ORDER BY Listing_Count ASC;
```

> There is an abundance of listings that are in the Manhattan and Brooklyn boroughs. There could be several reasons for this.

## **Distribution of Room Types**

```sql
SELECT Room_Type, COUNT(*) AS Room_Count
FROM [AirBnB in NYC]
GROUP BY Room_Type;
```

> There are 17,020 entire home/apartment listings in this dataset, constituting the majority of the room types. This category represents 55% of the total listings. This observation suggests that Airbnb hosts may prefer listing entire homes/apartments due to the potential for higher profits and greater convenience for renters


## **Average Price by Room Type**

```sql
SELECT Room_Type, AVG(Price) AS Average_Price
FROM [AirBnB in NYC]
GROUP BY Room_Type;
```

> The entire home/apartment room type has the highest average price, confirming the hypothesis that these listings possess the highest average and might be the reason why it is the most common listing type.

## **Average Price by Borough**

```sql
SELECT Borough, AVG(Price) AS Average_Price
FROM [AirBnB in NYC]
GROUP BY Borough;
```

> This query aids in establishing the average price for each borough. It provides insights into which borough boasts the highest-priced Airbnb listings and whether this correlates with reviews. The proximity of each borough to popular sites or tourist attractions could be a factor attracting renters to choose Airbnb rentals in these areas.

## **Average Review Score by Borough**

```sql
SELECT Borough,AVG(Review_Scores_Rating) AS Average_Review_Score
FROM [AirBnB in NYC]
GROUP BY Borough
ORDER BY Average_Review_Score DESC;

```

> The difference in the average review score among different boroughs is minimal. This indicates that the borough itself does not exert a significant influence on the review scores.

## **Are the Ratings Correlated to the Price?**

```sql
SELECT CAST(Price AS FLOAT) AS Price,
       CAST(Review_Scores_Rating AS FLOAT) AS Review_Scores_Rating
FROM [AirBnB in NYC]
WHERE Price IS NOT NULL
AND Review_Scores_Rating IS NOT NULL
AND ISNUMERIC(Price) = 1
AND ISNUMERIC(Review_Scores_Rating) = 1;
```

> The scatterplot reveals that, despite numerous listings receiving high ratings, the average price remains low. This implies that the rating does not influence the cost of each listing significantly. There are only a few outlier listings.

## **Number of Hosts/Listings per Year**

```sql
SELECT YEAR(Host_Since) AS Hosting_Year, COUNT(DISTINCT Host_ID) AS Host_Count
FROM [Airbnb in NYC]
GROUP BY YEAR(Host_Since)
ORDER BY Hosting_Year;
```

> 
The line chart illustrates the decline in the number of hosts joining Airbnb. While examining the dataset, a substantial number of blanks were observed. 62% of these blanks originated from January to December 2015. A law, specifically the New York State dwelling law, was enacted, prohibiting full-apartment rentals for less than 30 days. This empowers authorities to impose fines of up to $7,500 on hosts caught with such listings.
